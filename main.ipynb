```python
from sportsipy.mlb.teams import Teams
import matplotlib.pyplot as plt
```

```python
def generate_visualization(year, winner):    # NOTE: This years thing should be 
    name, obp, era = ([], [], [])

    for team in Teams(year)._teams:
        team_obj = Teams(year)[team.abbreviation]

        if team_obj.rank <= 10:
            name.append(team.abbreviation)
            obp.append(team_obj.on_base_percentage)
            era.append(team_obj.earned_runs_against)

    for i, label in enumerate(name):
        if label == winner:
            plt.annotate(f'{label} - {year}', (obp[i], era[i]))   
            plt.scatter(obp[i], era[i], c='green') 
        elif winner == '?':    # In this case, it's 2022 season... so there's no winner
            plt.annotate(label, (obp[i], era[i]))
            plt.scatter(obp[i], era[i], c='blue')
        else:
            plt.scatter(obp[i], era[i], c='red')
    
    plt.xlabel('OBP')
    plt.ylabel('ERA')
```

```python
generate_visualization(2000, winner='NYY')
generate_visualization(2001, winner='ARI')
generate_visualization(2002, winner='LAA')
generate_visualization(2003, winner='MIA')
generate_visualization(2004, winner='BOS')
generate_visualization(2005, winner='CHW')
generate_visualization(2006, winner='STL')
generate_visualization(2007, winner='BOS')
generate_visualization(2008, winner='PHI')
generate_visualization(2009, winner='NYY')
generate_visualization(2010, winner='SFG')
generate_visualization(2011, winner='STL')
generate_visualization(2012, winner='SFG')
generate_visualization(2013, winner='BOS')
generate_visualization(2014, winner='SFG')
generate_visualization(2015, winner='KC')
generate_visualization(2016, winner='CHC')
generate_visualization(2017, winner='HOU')
generate_visualization(2018, winner='BOS')
generate_visualization(2019, winner='WSN')

plt.title(f'Regular Season OBP vs ERA (2000-2019)')


left, right = plt.xlim()
bottom, top = plt.ylim()
```

```python
generate_visualization(2022, winner='?')
plt.title(f'Regular Season OBP vs ERA 2022')
plt.xlim(left, right)
plt.ylim(2.75, top)
```
